import pandas as pd

⸻

Step 2: Select Numerical Columns

Logic:
	•	Only grab the numeric columns (exclude names, IDs, categories).
	•	Look at .info() or .head() to double-check.

numeric_data = data[['NumEmployees', 'Revenue', 'Expenses', 'NumAisles', 'SquareFootage', 'DailyCustomers', 'ParkingSpaces']]



⸻

Step 3: Scale the Data (Important)

Logic:
	•	Standardize the data so that features are on the same scale.
	•	PCA assumes features are centered and scaled.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)



⸻

Step 4: Run PCA

Logic:
	•	Fit the PCA model.
	•	Check eigenvalues (variance explained by each component).

pca = PCA()
pca.fit(scaled_data)

# View eigenvalues
print(pca.explained_variance_)

Interpretation Tip:
Look for eigenvalues > 1 (Kaiser rule) to decide how many components to keep.

⸻

Step 5: Scree Plot

Logic:
	•	Visualize how much variance each principal component explains.
	•	Look for the “elbow” where adding more components stops giving big benefits.

plt.plot(range(1, len(pca.explained_variance_) + 1), pca.explained_variance_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()

Interpretation Tip:
Pick the number of components before the curve flattens out.
